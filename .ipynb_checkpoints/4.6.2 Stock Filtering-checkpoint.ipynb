<br> 스코어 별로 최대 손실은 어떤지도 궁금합니다. 스코어가 높다고 무조건 매수했다가 큰 낙폭으로 손해를 볼 수도 있기 때문입니다. 

```python
X_outcome = pd.read_pickle('X_outcome.pkl')
X_outcome.head()
```

<br> 각 스코어 구간이 종가 수익율 구간에 따라 최고 수익율이 어떻게 변하는 지 보겠습니다. 결과를 보니 종가 수익률이 높거나 낮은 구간에서 더 좋은 수익율을 기대할 수 있습니다.

```python
X_outcome['yhat_rank'] = pd.qcut(X_outcome['yhat'], q=5)
X_outcome['return_rank']  = pd.qcut(X_outcome['return'], q=5)
X_outcome.groupby(['yhat_rank','return_rank'])['target'].mean().unstack()
```

<br> 참고로 groupby 로 데이터를 요약하는 방법은 직관적이나, 각 행과 열의 총계는 보여주지 않는다는 단점이 있습니다. 총계가 보고 싶을 때는 pivot_table 를 아래와 같이 이용합니다.

```python
pd.pivot_table(data = X_outcome, index = 'yhat_rank', columns = 'return_rank', values = 'max_close', aggfunc='mean', margins=True)
```

<br> 각 스코어 구간이 종가 수익율 구간에 따라 최고 수익율이 어떻게 변하는 지 보겠습니다. 스코어 구간이 낮거나 높은 구간에서는 최저 수익률도 낮습니다. 즉 변동성이 아주 큰 구간이라고 판단할 수 있습니다. 수익률이 비교적 낮더라도, 큰 리스크를 앉고 매매를 하고 싶지는 않습니다. 따라서, 우리가 매수할 종목은 스코어는 

```python
X_outcome.groupby(['yhat_rank','return_rank'])['max_close'].min().unstack()
```
