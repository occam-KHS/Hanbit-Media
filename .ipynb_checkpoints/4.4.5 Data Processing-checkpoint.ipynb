### <br>전략 1 - 모든 종목 매수 후, 5 영업일 기간 6% 익절 6% 손절
한가지 전략을 테스트 해 보겠습니다. 모든 종목을 같은 금액으로 매일 종가 매수합니다. 매수 후 5 영업일 동안 6% 익절, 6% 손절합니다. 익절/손절외 종목은 5 영업일에 전부 종가 매도하면 수익율은 어떻게 될까요?   

```python
kosdaq_list = pd.read_pickle('kosdaq_list.pkl')

data_all_1 = pd.DataFrame()

ub = 1.06
lb = 0.94

for code in kosdaq_list['code']:
    
    # 종목별 처리
    data = mdl_data[mdl_data['code']==code].sort_index().copy()
    
    # 최고/최저 수익율
    for i in [1,2,3,4,5]:

        data['high_r' + str(i)] = data['high'].shift(-1*i)/data['close']      
        data['low_r' + str(i)] = data['low'].shift(-1*i)/data['close']   
        data['close_r' + str(i)] = data['close'].shift(-1*i)/data['close']    
        
    data['profit_sell']  = (data[['high_r1','high_r2','high_r3','high_r4','high_r5']].max(axis=1) > ub).astype(int)
    data['loss_sell']    = (data[['low_r1','low_r2','low_r3','low_r4','low_r5']].min(axis=1) < lb).astype(int)                 
        
    data['final_return'] = np.where(data['profit_sell']==1, ub, np.where(data['loss_sell']==1, lb, data['close_r5']))
    data['final_return'] = np.where((data['profit_sell']==1) & (data['loss_sell']==1), np.nan, data['final_return']) # 익절/손절 둘다 해당하면 수익율 계산에서 제외
       
    data.dropna(subset=['close_r1','close_r2','close_r3','close_r4','close_r5'], inplace=True)   
    data_all_1 = pd.concat([data, data_all_1], axis=0)

data_all_1.to_pickle('data_all_1.pkl')    
data_all_1.head()    
```

### <br>전략 2 - 모든 종목을 종가 매수 후, 아래와 같은 순서로 매도  
1. (익일 고가/매수 종가) 값이 (당일 고가/ 매수 종가) 값 보다 크면 익일 종가 매도   
2. (당일 고가/매수 종가) 값이 1.06 이면 매도   
3. 매도 안된 종목은 전부 5 영업일 날 종가 매도  


위와 같은 매도 전략은 수익율이 어떻게 될까요?   

```python
def final_r(x):
    
    if ((x['high']/x['close']) < x['high_r1']):        
        return x['close_r1']
    
    elif x[['high_r1','high_r2','high_r3','high_r4','high_r5']].max() > ub:
        return ub
    
    else:
        return x['close_r5']        
    
for code in kosdaq_list['code']:    
    
    # 종목별 처리
    data = mdl_data[mdl_data['code']==code].sort_index().copy()
    
    # 최고/최저 수익율
    for i in [1,2,3,4,5]:

        data['high_r' + str(i)] = data['high'].shift(-1*i)/data['close']      
        data['low_r' + str(i)] = data['low'].shift(-1*i)/data['close']   
        data['close_r' + str(i)] = data['close'].shift(-1*i)/data['close']
        
    data['final_return'] = data.apply(final_r, axis=1)
                                                                                                                                                 
    data.dropna(subset=['close_r1','close_r2','close_r3','close_r4','close_r5'], inplace=True)   
    data_all_2 = pd.concat([data, data_all_2], axis=0)

data_all_2.to_pickle('data_all_2.pkl')    
data_all_2.head()    
```

### <br>전략 3 - 래리 윌리암스의 변동성 돌파 전략의 응용.       
래리 윌리암스의 변동성 돌파 전략은 다음날 시가에 전부 매도하는 것입니다. 만약 매수 후 5 영업일동안 익절/손절 6% 라인으로 매도하고, 매도가 안 되면 5 영업일에 전부 종가로 매도해 보겠습니다.   
위와 같은 매도 전략은 수익율이 어떻게 될까요?

```python
kosdaq_list = pd.read_pickle('kosdaq_list.pkl')

data_all_3 = pd.DataFrame()

ub = 1.06
lb = 0.94

for code in kosdaq_list['code']:
    
    # 변동성 돌파전략
    data = mdl_data[mdl_data['code']==code].sort_index().copy()
    
    data['v'] =  (data['high'].shift(1) - data['low'].shift(1))*0.6   
    data['buy_price'] = data['open'] + data['v']
    data['buy'] = np.where((data['low'] < data['buy_price']) & (data['high'] > data['buy_price']), 1, 0)
    data.dropna(subset=['v','buy_price'], inplace=True) # Missing 데이터 제거    
    
    # 최고/최저 수익율
    for i in [1,2,3,4,5]:

        data['high_r' + str(i)] = data['high'].shift(-1*i)/data['buy_price']      
        data['low_r' + str(i)] = data['low'].shift(-1*i)/data['buy_price']   
        data['close_r' + str(i)] = data['close'].shift(-1*i)/data['buy_price']    
        
    data['profit_sell']  = (data[['high_r1','high_r2','high_r3','high_r4','high_r5']].max(axis=1) > ub).astype(int)      
    data['loss_sell']    = (data[['low_r1','low_r2','low_r3','low_r4','low_r5']].min(axis=1) < 0.94).astype(int)   

    data['final_return'] = np.where(data['profit_sell']==1, ub, np.where(data['loss_sell']==1, lb, data['close_r5']))
    data['final_return'] = np.where((data['profit_sell']==1) & (data['loss_sell']==1), np.nan, data['final_return']) # 익절 손절 둘다 해당하면 수익율 계산에서 제외
    
    data.dropna(subset=['close_r1','close_r2','close_r3','close_r4','close_r5'], inplace=True)   
    
    data_all_3 = pd.concat([data, data_all_3], axis=0)
    
data_all_3.to_pickle('data_all_3.pkl')    
data_all_3.head()      
```

### <br>전략 2 - 평균회귀   
이번에 이동 평균선을 이용한 전략을 시도해 보겠습니다. 전 일 이동평균선(5일)이 종가보다 높다면, 주가가 추세보다 낮게 형성되어 있다고 생각할 수 있을 것 같습니다. 당일 전 일 종가보다 1.5% 빠졌을 때 즉시 매수합니다. 그리고 전략 1 과 동일한 매도 전략을 취해보겠습니다.

```python
kosdaq_list = pd.read_pickle('kosdaq_list.pkl')

data_all_4 = pd.DataFrame()

ub = 1.06
lb = 0.94

for code in kosdaq_list['code']:
    
    # 매수 가격 결정
    data = mdl_data[mdl_data['code']==code].sort_index().copy()
    data['ma_5d'] = data['close'].rolling(5).mean()
    data['buy_signal'] =  (data['close'].shift(1) < data['ma_5d'].shift(1)).astype(int) # 전일 종가가 5일 이동평균선보다 낮다.   
    data['buy'] = (data['buy_signal']==1)*(data['low'] < data['close'].shift(1)*0.985).astype(int) # 전일 종가가 5일 이동평균선보다 낮고, 전일 종가보다 1.5% 빠졌을 때 매수
    data['buy_price'] = data['close'].shift(1)*0.985 # 매수가
    data.dropna(subset=['ma_5d','buy_signal','buy','buy_price'], inplace=True) # Missing 데이터 제거    
    
    # 최고/최저
    for i in [1,2,3,4,5]:

        data['high_r' + str(i)] = data['high'].shift(-1*i)/data['buy_price']      
        data['low_r' + str(i)] = data['low'].shift(-1*i)/data['buy_price']   
        data['close_r' + str(i)] = data['close'].shift(-1*i)/data['buy_price']    
        
    data['profit_sell']  = (data[['high_r1','high_r2','high_r3','high_r4','high_r5']].max(axis=1) > ub).astype(int)
    data['loss_sell']    = (data[['low_r1','low_r2','low_r3','low_r4','low_r5']].min(axis=1) < lb).astype(int)   
    
            
    data['final_return'] = np.where(data['profit_sell']==1, ub, np.where(data['loss_sell']==1, lb, data['close_r5']))
    data['final_return'] = np.where((data['profit_sell']==1) & (data['loss_sell']==1), np.nan, data['final_return']) # 익절 손절 둘다 해당하면 수익율 계산에서 제외       
        
    data.dropna(subset=['close_r1','close_r2','close_r3','close_r4','close_r5'], inplace=True) 
    
    data_all_4 = pd.concat([data, data_all_4], axis=0)
```

- #### 결과기간의 수익율

익절과 손절라인에 따른 매도전략을 실행하였을 때, 수익율이 어떻게 될 지 계산해 보겠습니다.
아래 코드에서 ###### 수익율 계산 ####### 아래에 해당합니다. 수익율을 계산하는 로직을 만드는 방법은 여러가지가 있을 것입니다. 구현된 코드는 여러가지 로직 중 하나입니다. 

outcome_all 데이터프레임에서 종목 리스트을 만든 후 for loop 으로 한 종목씩 데이터를 처리합니다. code_df 는 종목별 날짜별로 결과가 담겨있는 데이터프레임입니다. 혹시 날짜 순으로 안 되어 있을 경우를 위해 sort_index() 한 번 해 줍니다. 그리고 profit_sell (고가 기준)이 발생한 날, loss_sell (종가 기준) 이 발생한 날과 마지막 날의 수익율을 results 에 담습니다. 이 세가지 이벤트 중, 가장 먼저 일어날 이벤트를 수익율로 결정할 것입니다. Profit_sell 은 장중에 발생한 고가 기준이고 loss_sell 은 종가 기준이므로 두 이벤트가 동시에 발생하면 profit_sell 이 먼저 발생한 이벤트라고 할 수 있을 것 입니다. 일봉 데이터의 경우, 고가와 저가 중 어느 것이 먼저 발생한 이벤트인지 알 수 없으므로, 순서를 알 수 있는 고가와 종가를 활용합니다.  

drop_duplicates 함수를 이용하여 동일한 날짜와 동일한 코드에 대하여 첫번째 이벤트만 results_all 에 저장합니다. results_all 는 일별 종목별 수익율을 가지고 있는 데이터프레임입니다. 

가설 검증할 데이터를 월별로 나누어 저장합니다. 가설이 시간에 상관없이 성립되는 지 알고 싶기 때문입니다.

```python
def strategy_simul(start_date, end_date):

    w_days = list(pd.bdate_range(start_date, end_date)) # 백테스트할 날짜모음

    results_all = pd.DataFrame()

    for dt in w_days: # 날짜를 하루 하루 증가시키면서 시뮬레이션 


        obs_end = datetime.datetime.strftime(dt, "%Y-%m-%d")
        start_date = datetime.datetime.strptime(obs_end, "%Y-%m-%d") - datetime.timedelta(days=180) # 총 180 일 관찰기간
        obs_start = datetime.datetime.strftime(start_date, "%Y-%m-%d")      

        code_list = []
        
        price_list = []
        price_mean_list = []
        price_std_list = []
        price_z_list = []
        price_z_out_list = []
        price_5ma_list = []
        
        volume_list = []
        volume_mean_list = []
        volume_std_list = []
        volume_z_list = []
        volume_z_out_list = []
        
        wins_60_list = []
        wins_180_list = []

        toptail_list = []
        
        for code in kosdaq_list['code']:
            s = return_all[return_all['code']==code].loc[obs_start:obs_end]

            if (s['volume'].sum() < 1000000) or (s['close'].std() == 0): # 6 개월간 총 거래량이 백만은 넘겠지...
                continue

            code_list.append(code)

            price_list.append(s['close'].tail(1).max()) # 마지막날 종가
            price_mean_list.append(s['close'].mean()) # 종가 평균
            price_std_list.append(s['close'].std()) # 종가 표준편차
            price_z =  (s['close'].tail(1).max() - s['close'].mean()) / s['close'].std() # 어제 Z 값
            price_z_out =  np.where( ((s['close'] - s['close'].mean()) / s['close'].std()) > 1.6, 1, 0) # 특이하게 높은 값 기록
            price_5ma_list.append(s['close'].tail(5).mean()) # 5일 이동평균선

            volume_list.append(s['volume'].tail(1).max()) # 마지막날 거래량
            volume_mean_list.append(s['volume'].mean()) # 거래량 평균
            volume_std_list.append(s['volume'].std()) # 거래량 표준편차
            volume_z = (s['volume'].tail(1).max() - s['volume'].mean()) / s['volume'].std()  # 어제 Z 값
            volume_z_out = np.where( ((s['volume'] - s['volume'].mean()) / s['volume'].std()) > 1.6, 1, 0) # 특이하게 거래량 높은 날 기록

            price_z_list.append(price_z)
            volume_z_list.append(volume_z)

            price_z_out_list.append(price_z_out.sum())
            volume_z_out_list.append(volume_z_out.sum())

            wins_60_list.append(s['win_market'].tail(60).sum()) # 지난 180일 동안 코스닥 인덱스보다 잘 한 날짜 수
            wins_180_list.append(s['win_market'].sum()) # 지난 60일 동안 코스닥 인덱스보다 잘 한 날짜 수
            
            toptail = (s['close'] > s['open'])*((s['high']/s['close'])>1.05).astype('int') # 양봉이고 5% 이상 위 꼬리 상승
            toptail_list.append(toptail.sum())
            

        # 데이터프레임으로 전환 - 종목레벨 데이터
        wins = pd.DataFrame({'code': code_list, 'price': price_list, 'price_mean': price_mean_list, 'price_std': price_std_list, 'price_z': price_z_list,'price_z_out': price_z_out_list, 'price_5ma': price_5ma_list,\
                             'volume': volume_list, 'volume_mean': volume_mean_list, 'volume_std': volume_std_list, 'volume_z': volume_z_list, 'volume_z_out': volume_z_out_list,\
                             'num_wins_60': wins_60_list, 'num_wins_180': wins_180_list, 'num_toptail': toptail_list})


        wins['buy_price'] = wins['price']*0.985 # 매수 가격은 마지막날 종가에서 1.5% 빠진 가격

        c1 = (wins['num_wins_180'] > 25)  # 지난 180일 동안 코스닥 인덱스보다 잘 한 날짜 수 > 20 (한달 영영일)
        candidates = wins[ c1 ] # 위 조건을 만족
        

        out_start =  datetime.datetime.strftime(datetime.datetime.strptime(obs_end, "%Y-%m-%d") + datetime.timedelta(days=1), "%Y-%m-%d") # 구매시작일
        out_end  =  datetime.datetime.strftime(datetime.datetime.strptime(obs_end, "%Y-%m-%d") + datetime.timedelta(days=7), "%Y-%m-%d") # 보유하는 날짜

        profit_sell = 1.08 # 익절라인
        loss_sell = 0.96 # 손절라인

        outcome_all = pd.DataFrame()   
       
        for index, row in candidates.iterrows():

            outcome = return_all[return_all['code']==row['code']].loc[out_start : out_end][['code', 'low', 'high', 'close']] # 필요한 컬럼만
            outcome['buy_price'] = int(row['buy_price']) # 어제 종가
            outcome['highest'] = outcome['high'].max()
            outcome['lowest'] = outcome['low'].min()
            outcome['price_z'] = row['price_z'] # 어제 정보
            outcome['volume_z'] = row['volume_z'] # 어제 정보
            outcome['price_z_out'] = row['price_z_out'] # 180일 요약 정보
            outcome['price_5ma'] = row['price_5ma'] # 5일 요약 정보
            outcome['volume_z_out'] = row['volume_z_out'] # 180일 요약 정보
            outcome['num_wins_60'] = row['num_wins_60'] # 60일 요약 정보
            outcome['num_wins_180'] = row['num_wins_180'] # 180일 요약 어제 정보    
            outcome['num_toptail'] = row['num_toptail'] # 180일 요약 어제 정보    
            outcome['num_wins_trend'] = outcome['num_wins_60'] / outcome['num_wins_180']
            outcome['price_from_5ma'] = outcome['buy_price'] / outcome['price_5ma']
            outcome['yymmdd'] = obs_end
            
            outcome['buy'] = np.where(outcome['low'].head(1).max() < outcome['buy_price'].max(), 1, 0) # 당일 저가가 매수할 가격보다 내려갔나?
            
            outcome['return_close'] = outcome['close'] / outcome['buy_price'] # 종가 수익율: (오늘 종가 / 매수가격)
            outcome['return_high'] = outcome['high'] / outcome['buy_price'] # 최고가 수익율: 최고가 / 매수가격)   
            outcome['return_low'] = outcome['low'] / outcome['buy_price'] # 최저가 수익율: 최저가 / 매수가격)                                                                                                                                                                                    
            
            outcome['profit_sell'] = np.where(outcome['return_high'] > profit_sell, 1, 0)  # 고가 수익율이 악절라인 도달?
            outcome['loss_sell'] = np.where(outcome['return_close'] < loss_sell, 1, 0)  # 종가 수익율이 손절라인 도달?     
                                                 
            outcome_all = pd.concat([outcome_all, outcome], axis=0)         
                                          
                  
        ###############################     수익율 계산    ################################# 
        
        for code in list(outcome_all['code'].unique()):   
            code_df = outcome_all[outcome_all['code']==code].sort_index()
            profit = code_df[code_df['profit_sell']==1].head(1)  # 고가 기준으로 익절한 첫번째 이벤트
            loss = code_df[code_df['loss_sell']==1].head(1) # 종가기준으로 손절한 첫번째 이벤트
            final = code_df.tail(1) # 마지막 날 이벤트  
            results = pd.concat([profit, loss, final], axis=0).sort_index() # 날짜 순으로 정렬

            results_all = pd.concat([results_all, results], axis=0) # 전체 이벤트 합함
            results_all = results_all.sort_index().drop_duplicates(subset=['yymmdd','code'], keep='first')  # 매일 발생한 이벤트 중 첫번째 이벤트                                
    
    return outcome_all, results_all
                                            
outcome_202105, results_202105 = strategy_simul('2021-05-01', '2021-05-31')
results_202105.to_pickle('results_202105.pkl')

outcome_202106, results_202106 = strategy_simul('2021-06-01', '2021-06-30')
results_202106.to_pickle('results_202106.pkl')

outcome_202107, results_202107 = strategy_simul('2021-07-01', '2021-07-31')
results_202107.to_pickle('results_202107.pkl')

outcome_202108, results_202108 = strategy_simul('2021-08-01', '2021-08-31')
results_202108.to_pickle('results_202108.pkl')

outcome_202109, results_202109 = strategy_simul('2021-09-01', '2021-09-24')
results_202109.to_pickle('results_202109.pkl') 
```
